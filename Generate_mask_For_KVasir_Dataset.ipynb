**Utility function to inpaint on image dataset**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import os

def inpaint_image(inpaint_dir, mask_dir):

  mask = cv2.imread(mask_dir)
  mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)

  for filename in os.listdir(inpaint_dir):
    img = cv2.imread(os.path.join(inpaint_dir, filename))
    dst = cv2.inpaint(img, mask, 100, cv2.INPAINT_TELEA)
    cv2.imwrite(inpaint_dir+'/'+filename, dst)
```
